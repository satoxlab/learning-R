Rの基本


#######################################################


代入


Rでは、#のあとの文章は、リターンキーで改行されるまで「無視」されます。


そのため、エディタ（スクリプト）ファイル上にメモを保存しておきたいとき、#を付してから文章を入力しておくと便利です。


不等号とマイナスの組み合わせ <- は「矢印」を意味します。矢印の方向に、代入するという意味です。


```r
c(1, 2, 3) -> kazu
```

例えば、これは1, 2, 3という数列をkazuという文字に代入しています。確かめてみましょう。


```r
kazu
```

こんどは矢印を逆向きにします。


```r
kazu2 <- c(1, 2, 3)  
kazu2
```

同じように代入できましたね。


ダブルクオテーション「 " 」は数字以外のデータの場合に使います。

```r
namae <- c("taro", "hanako", "犬")  
```

実際のデータ解析ではデータの名前を xとかyなどとするよりも、sizeとかkazuなど具体的なnamaeなどの方が分かりやすいです。


またRでは大文字と小文字を「違うもの」として区別します。だから、namaeとNamaeに別のデータを代入できます。


```r
Namae <- c("cat", "dog", "fish")
namae  
Namae
```

なお、コンマやカッコの後などに半角スペースを入れなくてもかまいません。


半角スペースを入れておかないと、あとで呪文が読みにくいという欠点がありますが、入力は速くなります。半角スペースを入れずに入力すると、


```r
c(4,5,6)->kazu
```

例えばこうなります。


```r
kazu
```

数列


こんどは等差数列をつくります。


```r
seq(1, 4, 0.5) 
```

seqはsequenceの意味です。上記は1から4までの0.5間隔での数列です。


```r
3:1
```

これはseq(3, 1, 1)と同じです。コロン「 : 」は、等差数列の呪文seqの簡略型です。


```r
seq(3, 1, -1)
```

```r
seq(1,5,1)
```

したがって、1:5も、seq(1, 5, 1)のことです。今度はrepです。


```r
rep("a", 4) 
```

 repはreplicateの意味です。ここでは"a"を4つ複製します。

```r
rep(3, 10) 
```

同様に、これは3を10個複製するという呪文です。


```r
z <- c(rep(3, 4), rep(c(1, 5, 10), c(2, 3, 4)))
z
```

なお、？を頭につけて呪文を唱えると、help画面を呼び出すことができます。


```r
?rep
```

Rを使った作図の基本


#######################################################


１種類の量的データ（例：1.3, 2.5, 1.2, 1.4, ...）を作図する


データ例


ここでは、irisというRの付録データ (アヤメ (花) の部位と種名のデータ) を使います。

まず、irisを呼び出します。


```r
data(iris)  
```

次にirisがどんなデータか、summaryを見てみましょう。

```r
summary(iris) 
```

Sepal Lengthは、がくの長さ、Sepal Widthは、がくの幅、Petal Lengthは花びらの長さ、Petal Lengthは花びらの幅、Speciesは種名です。


```r
attach(iris)
```

Rにirisを取り付けます。これで、以後iris$を唱えなくても大丈夫です。


ヒストグラム


ヒストグラムは、最も代表的な作図方法です。


```r
hist(Sepal.Length)
```

```r
hist(Petal.Length)
```

histで作図すると、データの頻度分布を把握できます。

箱ひげ図

もうひとつの代表的な方法として、箱ひげ図があります。

```r
boxplot(Sepal.Length)
```

箱の中にある太い線は中央値、箱の上下が四分位数、ひげの末端が最小値、最大値です。

ただし、頻度分布が複数の山の形になっているときなどは、箱ひげ図は不適切です。

```r
boxplot(Petal.Length)
```

histで見たように、本当は２つの山があるのに、箱ひげ図では、そのことが分かりません。

軸の指定


X軸とY軸の下限値と上限値は、それぞれxlimとylimで指定できます。

```r
hist(Sepal.Length, xlim=c(0, 10), ylim=c(0, 50))
```

軸に名称を付ける場合は、xlabやylabを使います。

```r
boxplot(Sepal.Length, ylim=c(0, 10), xlab="Iris", ylab="Sepal Length")
```

また、ヒストグラムの間隔は、breaksによって指定できます。

```r
hist(Sepal.Length, xlim=c(0, 10), ylim=c(0, 60), breaks=0:10)
```

attachしたデータの取り外します。

```r
detach(iris)
```

Rを使った確率分布の基本

#######################################################


正規分布

確率分布のコマンドは、頭文字１字で「Rがすべき作業内容」を表し、２文字目から確率分布の略称となっていることが多いです。


正規分布の英名はnormal distributionなので「norm」が正規分布の略称です。

頭文字1字には、d、p、rの３通りがあります。それぞれの作業内容は以下のとおりです。

 dnorm: 想定した正規分布におけるX軸の値ごとの「確率密度」を求める。densityのd

 pnorm: 想定した正規分布において、X軸が、ある値以下となる(orより大きな値となる)「確率」を求める。probabilityのp

rnorm: 想定した正規分布から、X軸の値をランダムに選択する。randomのr

確率密度

ある正規分布における各値の確率密度を計算したい場合はdnormを用います。

母平均が1で母分散が4（母標準偏差が2）の正規分布で、xが3になる確率密度を求めましょう。


```r
dnorm(3, mean=1, sd=2)
```

ちょっと違う値も試してみましょう。

```r
dnorm(1, mean=1, sd=2)
```

```r
dnorm(1.1, mean=1, sd=2)
```

```r
dnorm(2, mean=1, sd=2)
```

では、値が3以下になる確率はなんでしょうか。

```r
pnorm(3, mean=1, sd=2)
```

結構大きいですね。lower.tail=TRUEだからです。

lower.tail=FALSEにすると、Xが3以上の面積を計算します。

```r
pnorm(3, mean=1, sd=2, lower.tail=F) 
```

```r
pnorm(3, mean=1, sd=2)+pnorm(3, mean=1, sd=2, lower.tail=F) 
```

両方を足すと１になりましたね。どんな確率分布でも、確率(密度)分布の総面積は１です。

平均0で分散 (標準偏差) 1の正規分布を標準正規分布といいます。

二項分布

確率分布のコマンドは、頭文字１字で「Rがすべき作業内容」を表し、２文字目から確率分布の略称となっていることが多いです。


二項分布の英名はbinomial distributionなので「binom」が二項分布の略称です。


頭文字1字には、d、p、rの３通りがあります。それぞれの作業内容は以下のとおりです。


dbinom: 想定した二項分布におけるX軸の値ごとの「確率」を求める。densityのd


 pbinom: 想定した二項分布において、X軸が、ある値以下となる(orより大きな値となる)「確率」を求める。probabilityのp


rbinom: 想定した二項分布から、X軸の値をランダムに選択する。randomのr


ある値となる確率

ある二項分布における各値の確率を計算したい場合はdbinomを用います。

母比率probが0.5で試行回数sizeが10の二項分布で、xが3になる確率密度を求めましょう。

```r
dbinom(3, size=10, prob=0.5)
```

別の値も試してみましょう。

```r
dbinom(5, size=10, prob=0.5)
dbinom(7, size=10, prob=0.5)
dbinom(10, size=10, prob=0.5)
```

ある値以下 (よりも大きな値) となる確率

では、値が3以下になる確率はなんでしょうか。

```r
pbinom(3, size=10, prob=0.5)  
```

```r
?pbinom 
```

このヘルプ画面を読むと、pbinomではlower.tail=TRUEになっています。

つまり、ここでは、x=3, 2, 1, 0となる確率の和を計算しています。

```r
dbinom(3, size=10, prob=0.5)+dbinom(2, size=10, prob=0.5)+dbinom(1, size=10, prob=0.5)+dbinom(0, size=10, prob=0.5)
```

lower.tail=FALSEにすると、Xが3よりも大きな値となる確率を計算します。

```r
pbinom(3, size=10, prob=0.5, lower.tail=F)  
```

```r
pbinom(3, size=10, prob=0.5) + pbinom(3, size=10, prob=0.5, lower.tail=F)  
```

両方を足すと１になりましたね。どんな確率分布でも、確率分布の総面積は１です。

母比率probが0.5で試行回数sizeが10の二項分布を描きましょう。

```r
x <- 0:10 
```

0から10までの整数の数列をつくります。

このxの値ごとで確率を計算して、それを線で結びます (type="l")。

注意：type="l"のlはイチではなくエルの小文字です。

```r
plot(x, dbinom(x, size=10, prob=0.5), type="l", ylim=c(0, 0.5))
```

いちど作図されたあとで線を描き足すにはlinesを使います。

母比率を変えて重ね書きしてみましょう。

二項分布は母比率が0あるいは1に近いほど左右非対称な形になります。

このように、二項分布は母比率と試行回数で形が決まります。

ｔ分布

ｔ分布はｔ検定を始めとする仮説検定をおこなうための確率分布です。

確率分布のコマンドは、頭文字１字で「Rがすべき作業内容」を表し、２文字目から確率分布の略称となっていることが多いです。

ｔ分布の英名はt distributionなので「t」が正規分布の略称です。

頭文字1字には、d、p、rの３通りがあります。ｔ分布ではrを滅多に使わないので、省略します。

 dt: 想定したｔ分布におけるX軸の値ごとの「確率密度」を求める。densityのd

 pt: 想定したｔ分布において、X軸が、ある値以下となる(orより大きな値となる)「確率」を求める。probabilityのp


確率密度

あるｔ分布における各値の確率密度を計算したい場合はdtを用います。

自由度が１のｔ分布で、ｔ値が0の確率密度を求めましょう。

```r
dt(0, df=1)
```

ｔ分布の中心はｔ値が０のときです。

```r
dt(-1, df=1)  
```

```r
dt(1, df=1)
```

ｔ値が－１のときと１のときの確率密度は同じ値です。

このことから、ｔ分布が左右対称のようだと察することができますね。

ｔ分布の作図

ｔ分布は自由度dfによって形が決まるので、dfの値を変えて比べてみましょう。

```r
x <- seq(-5, 5, 0.01)
plot(x, dt(x, df=1), type="l", ylim=c(0, 0.5))  
lines(x, dt(x, df=10), col="red")
lines(x, dt(x, df=100), col="blue")
```

type="l"の「l」はイチではなくエルの小文字です。

自由度が大きくなるほど、ｔ分布は標準正規分布に近づきます。

```r
x <- seq(-5, 5, 0.01)
plot(x, dt(x, df=1), type="l", ylim=c(0, 0.5))  
lines(x, dt(x, df=10), col="red")
lines(x, dt(x, df=100), col="blue")
lines(x, dnorm(x, mean=0, sd=1), lwd=3, col="darkgreen")
```

ｔ値がー２以下になる確率を求めてみましょう。

```r
pt(-2, df=1)
```

デフォルトがlower.tail=TRUEなので、ｔ値が-2以下となる確率Pを求めています。

pt(-2, df=10)
pt(-2, df=100)  

このように、ｔ値が同じでも自由度が異なるときはP値も異なります。

ｔ検定などではｔ値とP値だけを記す人が多いですが、自由度dfの値も記しておくべきです。

χ2分布

χ2分布はχ2検定を始めとする仮説検定をおこなうための確率分布です。

確率分布のコマンドは、頭文字１字で「Rがすべき作業内容」を表し、２文字目から確率分布の略称となっていることが多いです。

χ2分布の英名はchi-squared distributionなので「chisq」が分布の略称です。

頭文字1字には、d、p、rの３通りがあります。χ2分布ではrを滅多に使わないので、省略します。

dchisq: 想定したχ2分布におけるX軸の値ごとの「確率密度」を求める。densityのd

pchisq: 想定したχ2分布において、X軸が、ある値以下となる(orより大きな値となる)「確率」を求める。probabilityのp

確率密度

あるχ2分布における各値の確率密度を計算したい場合はdtを用います。

自由度が１のχ2分布で、それぞれのχ2値の確率密度を求めましょう。

```r
dchisq(1, df=1)
dchisq(2, df=1) 
dchisq(5, df=1)  
```

今度は自由度が3のχ2分布で、それぞれのχ2値の確率密度を求めましょう。

```r
dchisq(1, df=3)
dchisq(2, df=3) 
dchisq(5, df=3)
```

χ2分布は自由度に応じて変化することが分かります。 

χ2分布の作図

χ2分布は自由度dfによって形が決まるので、dfの値を変えて比べてみましょう。

```r
x <- seq(0.1, 10, 0.01)
plot(x, dchisq(x, df=1), type="l", ylim=c(0, 0.5))  
```

type="l"の「l」はイチではなくエルの小文字です。

```r
x <- seq(0.1, 10, 0.01)
plot(x, dchisq(x, df=1), type="l", ylim=c(0, 0.5))  
lines(x, dchisq(x, df=3), col="red")
lines(x, dchisq(x, df=5), col="blue")
```

χ2値が8よりも大きな値をとる確率を求めてみましょう。

```r
pchisq(8, df=1, lower.tail=F)  
```

デフォルトがlower.tail=TRUEなので、ここではFALSEに変えて、χ2値が8よりも大きな値となる確率Pを求めています。

```r
pchisq(8, df=3, lower.tail=F)  
pchisq(8, df=5, lower.tail=F)  
```

このように、χ2値が同じでも自由度が異なるときはP値も異なります。

χ2検定などではχ2値とP値だけでなく、自由度dfの値も必要です。

F分布

F分布はF検定や分散分析などの仮説検定をおこなうための確率分布です。

 確率分布のコマンドは、頭文字１字で「Rがすべき作業内容」を表し、２文字目から確率分布の略称となっていることが多いです。

F分布の英名はF distributionなので「f」が正規分布の略称です。

頭文字1字には、d、p、rの３通りがあります。F分布ではrを滅多に使わないので、省略します。

 df: 想定したF分布におけるX軸の値ごとの「確率密度」を求める。densityのd。

自由度degree of freedomのdfと同じ表記で紛らわしいですが、、。

 pf: 想定したF分布において、X軸が、ある値以下となる(orより大きな値となる)「確率」を求める。probabilityのp

確率密度

あるF分布における各値の確率密度を計算したい場合はdfを用います。

F分布は２つの自由度で形が決まるので、df1とdf2を指定しなければなりません。

df1が１、df2が20のF分布で、F値が3の確率密度を求めましょう。

df(3, df1=1, df2=20)  
df(3, df1=10, df2=20)    
df(3, df1=1, df2=200)

自由度に応じて確率密度が変化します。

F分布の作図

```r
x <- seq(0, 10, 0.01)
plot(x, df(x, df1=1, df2=20), type="l", ylim=c(0, 1))
```

ype="l"の「l」はイチではなくエルの小文字です。

```r
x <- seq(0, 10, 0.01)
plot(x, df(x, df1=1, df2=20), type="l", ylim=c(0, 1))  
lines(x, df(x, df1=5, df2=20), col="red")
lines(x, df(x, df1=50, df2=20), col="blue")
```

このように、F分布は左右非対称な分布であり、自由度によって形が決まります。

F値が3より大きな値になる確率を求めてみましょう。

```r
pf(3, df1=1, df2=20, lower.tail=F)
pf(3, df1=10, df2=20, lower.tail=F)  
pf(3, df1=1, df2=200, lower.tail=F)
```

このように自由度が変わるとP値も変わります。

F分布からP値を求めたときは、F値だけでなく自由度df1, df2の値も記しておくべきです。

ｔ検定

ｔ検定はｔ分布を利用した「１群あるいは２群の平均の検定」です。

ｔ分布のページも参照してください。

１群のｔ検定、対応２群のｔ検定　

データ例

ｔ検定は量的データを使った仮説検定で用います。

そこで、irisというRの付録データ (アヤメ (花) の部位と種名のデータ) を使います。 

まず、irisを呼び出します。

```r
data(iris)  
```

次にirisがどんなデータか、summaryを見てみましょう。

```r
summary(iris)  
```

Sepal Lengthは、がくの長さ、Sepal Widthは、がくの幅、Petal Lengthは花びらの長さ、Petal Lengthは花びらの幅、Speciesは種名です。


```r
attach(iris) 
```

Rにirisを取り付けます。これで、以後iris$を唱えなくても大丈夫です。

１群のｔ検定

「Sepal.Widthの平均が2である」という帰無仮説を検定します。

```r
t.test(Sepal.Width, mu=2)
```

解析結果のうち、t = 29.71がｔ値、df = 149が自由度です。

そして、p-value < 2.2e-16 がP値です。e-16とは10の-16乗を意味します。

つまり、Sepal.Widthの平均が2であるという帰無仮説は棄却されました。

「平均が2ではない」ということです。

このデータの平均は最後のmean of x に3.057333と記されています。 

対応2群のｔ検定

対応２群のｔ検定は1群のｔ検定と同じです。

「Sepal.Widthの平均はSepal.Widthの平均と同じである」という帰無仮説を検定します。

これらのデータは個体ごとのデータなので、各個体のSepal.LengthとSepal.Widthの差をとります。

```r
Sepal.Length - Sepal.Width -> D
```

そして、「Dの平均が０である」という帰無仮説を検定します。

```r
t.test(D, mu=0)
```

出力結果の見方は１群のｔ検定と同じです。

attachしたデータの取り外します。

```r
detach(iris)
```

独立2群のｔ検定

こんどは、正規分布からランダムに値をサンプリングした仮想データを使いましょう。

以下がわからない人は、確率分布の正規分布のページを参照してください。

```r
rnorm(20, mean=0, sd=1) -> x0
x0
rnorm(20, mean=1, sd=1) -> x1
x1
```

「x1とx0の平均は同じである」という帰無仮説を検定します。

```r
t.test(x0, x1)
```

出力結果の冒頭でWelch Two Sample t-testとありますね。

Rではデフォルトで、Welchのｔ検定をおこないます。

２群の分散が等しい場合に用いることのできるStudentのｔ検定をするには、

```r
t.test(x0, x1, var.equal=T)
```

分散 (var) が等しいこと (equal) が成立 (TRUE) しているということです。

これら2種類のｔ検定の結果を見比べると、自由度が異なることが分かります。

Welchのｔ検定は自由度を調節することによって分散が異なる2群の平均を比較しています。

自由度が違えばP値も変化することについては、ｔ分布のページを参照してください。

二項検定

二項検定は二項分布を利用した「１群の頻度の検定」です。

二項分布のページも参照してください。

全部で20回対戦して、14回勝ったとします。

「勝率は0.5である」という帰無仮説を検定します。

```r
binom.test(14, 20, p=0.5)
```

p-value = 0.1153がP値です。

このP値を二項分布から計算します。

```r
dbinom(14:20, size=20, prob=0.5)
```

これらの値を全部足します。

```r
sum(dbinom(14:20, size=20, prob=0.5))
```

両側検定なので、これを２倍にします。

```r
2*sum(dbinom(14:20, size=20, prob=0.5))
```

このように二項検定は二項分布から直接確率を計算しています。

 Fisherの正確確率検定

Fisherの正確確率検定は「２群以上の頻度の違いの検定」です。　　　　　　　　　　　　　　　　　　　　　　　　　　　

エビのいる場所を潮だまりで調べました。

魚がいない場所では、18個体が岩場にいて2個体が海藻に隠れていました。

魚がいる場所では、４個体が岩場にいて16個体が海藻に隠れていました。

```r
rbind(c(18, 2), c(4, 16)) -> x
x
fisher.test(x)
```

p-value = 1.664e-05がP値です。e-05とは10の-5乗を意味します。
